```python
#Importing required Libraries
from sklearn.datasets import load_iris
import pandas as pd
import seaborn as sns
```

```python
##Loading Iris Dataset using seaborn
Dataset = sns.load_dataset('iris')
print(Dataset)
```

```python
## Plotting Pair Plots between features to show seperation between 3 classes (setosa, versicolor, virginica)
sns.set(style='whitegrid', color_codes=True)
P_plots = sns.pairplot(Dataset, hue='species', markers=["s", "D", "o"], height=4)
```

#### Seaborn plotpair function has been used to draw plot pairs between the 4 features/variables present in Iris dataset [Sepal Length, Sepal Width, Petal Length, Petal Width]. *Pairplots provide relationship between each variable present in the dataset.*
From above plotpairs, following are the observations:-

1) **"Petal Width" and "Petal Lenth"** could be considered as **most prominent features** for the seperation of 3 flower classes (setosa, Versicolor, Virginica), followed by "Petal Length" and "Sepal Length".

2) In all the pairplots, it is clearly visible that it is **much easier to separate setosa with versicolor/virgnica** (linearly separable), however versicolor and virginica are not at all linearly separable.

3) **"Sepal length" and "Sepal Width" are the worst combination of features** which one can use for separation of three flower classes as all the three classes data are overlapped with each other.

4) **"Petal Length" can be used to separate Setosa Class**, we could have also selected "petal_width" however setosa flower data is not distributed evenly/properly.

5) To seperate the three flower classes, features should be used in the following order:
**Petal Length-------->Petal Width-------->Sepal Length----->Sepal Width**

6) Petal Length can be used to define the data range of 3 species:
Setosa Flowers: [0,2]
Versicolor: [2,6]
Virginca: [3.5, 9]

This refers to the notion that **Setosa flowers have smallest petal length, Versicolor flower has medium sized and Virginica flowers has the largest petal length.**
